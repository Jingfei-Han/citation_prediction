Firstly, we need interpret the plot:
+ Y-axis: the probability of the number of papers by each country of total paper
+ "NOT IN CCF" refers to [X,A\*; X,A; X,B; X,C; X,...]
+ "IN CCF" refers to [A,\_; B,\_; C,\_]
+ "NOT IN CORE" refers to [A,X; B,X; C,X]
+ "IN CORE" refers to [\_,A\*; \_,A; \_,B; \_,C; \_;L; \_,Australisian; \_,Not ranked; \_,Unranked; \_,National]

From the picture above, we can find:
+ For "NOT IN CCF": From 2011 to 2014, the number of papers published by China was decreasing.
+ For "IN CCF": From 2011 to 2014, the number of papers published by China was increasing.
+ For "NOT IN CORE": From 2006 to 2012, the number of papers published by Australia was decreasing but increasing from 2012. **[Stable]**
+ For "IN CORE": Stable

# Ratio of papers in particular label to all papers in a country [Abnormal]

```python
df_conference.pivot_table('con_id', index=['CCF_classification'], \
                          columns=['CORE_classification'], aggfunc=len, margins=True)\
[['A*','A','B','C','L','Australasian','Unranked','X', 'All']].fillna(0)
```

**Metrics:**
$$M(c,y,l)=\frac{1}{|V_{l}|}\sum_{i \in V_{l}} log(x_{cyi}+1)$$
$$G(c,y,l)=\frac{M(c,y,l)}{\sum_{j \in label} M(c,y,j)} $$

Where, $c$ denotes country, $y$ year and $l$ label.

$V_{l}$ denotes conferences set whose element belong to $l$ label. $x_{cyi}$ denotes the number of papers published by country $c$ to the conference $i$ in year $y$.

```python
def GM(country, year, label, df_tmp_paper):
    #df_tmp_paper = df_paper2[(df_paper2['country']==country) & (df_paper2['publish_year']==year)]
    #tot = df_tmp_paper.paper_id.count() # total number of papers.
    df_label = df_conference[df_conference['label'] == label]
    res = 0
    con_size = 0
    for con_id in df_label.con_id:
        cnt = df_tmp_paper[(df_tmp_paper['con_id']==con_id)].paper_id.count()
        #if cnt is smaller than 10, we think the data is abormal and ignore it.
        #if cnt <= 10:
        #    continue
        res += np.log(cnt + 1)
        con_size += 1
    #con_size = df_label.con_id.count()
    if con_size == 0:
        return 0
    return res * 1.0 / con_size

def GM_total(country,year):
    df_tmp_paper = df_paper2[(df_paper2['country']==country) & (df_paper2['publish_year']==year)]
    #tot = df_tmp_paper.paper_id.count() # total number of papers.
    res_list = []
    #label_set = df_conference.label.unique()
    label_set = ['A,A*','A,A','A,B','B,A*','B,A','B,B','B,C','C,A*','C,A','C,B','C,C','X,A*','X,A','X,B','X,C','B,X','C,X', 'X,Australasian']
    for label in label_set:
        res = GM(country, year, label, df_tmp_paper)
        res_list.append(res)
    df_res = DataFrame(res_list)
    df_res.columns = ['GM']
    df_res['label'] = label_set
    df_res['prob'] = df_res.GM / df_res.GM.sum()
    return df_res
```

```python
def get_trend(country):
    df_trend = DataFrame([])
    country = country
    for year in range(2000, 2016):
        df_res = GM_total(country,year)
        df_trend[year] = df_res['prob']
    df_trend.index = df_res['label']
    return df_trend

def get_GM(country):
    df_trend = DataFrame([])
    country = country
    for year in range(2000, 2016):
        df_res = GM_total(country,year)
        df_trend[year] = df_res['GM']
    df_trend.index = df_res['label']
    return df_trend
```

```python
def drawLM(df_trend, label_set,low_year, up_year,pre=True):
    if len(label_set) == 0:
        df_trend2 = df_trend
    else:
        df_trend2 = df_trend.ix[label_set]
    df_tmp = df_trend2.stack().reset_index()
    df_tmp.columns = ['label', 'year','prob']
    df_tmp = df_tmp[(df_tmp['year']>=low_year) & (df_tmp['year']<=up_year)]
    if pre==True:
        df_tmp['classification'] = df_tmp['label'].map(lambda x: x[0])
    else:
        df_tmp['classification'] = df_tmp['label'].map(lambda x: x.split(',')[1])
    #print df_tmp.head()
    g = sns.lmplot(x="year", y="prob", hue="label", col='classification', truncate=True, col_wrap=2, size=5, data=df_tmp)
    #return df_tmp
```

```python
df_trend1 = get_trend("China")
drawLM(df_trend1,[], 2011, 2015,True)
```
